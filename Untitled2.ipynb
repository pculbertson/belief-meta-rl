```python
import torch
import numpy as np
```

```python
class TestLSTM(torch.nn.Module):
    def __init__(self,input_size,hidden_size,output_size,num_layers,batch_first=True):
        super(TestLSTM, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.num_layers = num_layers
        self.LSTM = torch.nn.LSTM(input_size,hidden_size,num_layers,batch_first=batch_first)
        self.relu = torch.nn.ReLU()
        self.linear = torch.nn.Linear(hidden_size,output_size)
            
    def forward(self, x, hidden):
        x, hidden = self.LSTM(x,hidden)
        x = self.linear(self.relu(x))
        return x, hidden
```

```python
#count to 5
hidden = 10
batch = 5
layers = 2
input = 1
output = 1
seq = 10

lstm = TestLSTM(input,hidden,output,layers)
inputs = torch.from_numpy(np.array([0,1,2,3,4])).float().unsqueeze(1).expand(5,seq).unsqueeze(2)
h0=torch.zeros(layers,batch,hidden)
c0=torch.zeros(layers,batch,hidden)
outs= []
hidden = (h0,c0)
for i in range(seq):
    out, hidden = lstm(inputs[:,i,:].unsqueeze(2),hidden)
    outs.append(out)
outputs = torch.squeeze(torch.stack(outs,axis=1))
print(outputs.shape)
outs, hidden = lstm(inputs,(h0,c0))
print(outs.shape)
```

```python
out_des = torch.from_numpy(np.array([[i*n for i in range(0,10)] for n in range(1,6)])).float().unsqueeze(2)
torch.nn.MSELoss()(out,out_des)

opt = torch.optim.Adam(lstm.parameters(),lr=1e-1)
num_epochs = 1000
for _ in range(num_epochs):
    opt.zero_grad()
    out, hidden = lstm(inputs,(h0,c0))
    loss = torch.nn.MSELoss()(out,out_des)
    print(loss)
    loss.backward()
    opt.step()
```
